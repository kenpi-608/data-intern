# データ 前処理

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_table("ml-100k/u.data", header=None)
```

```python
df
```

```python
df_s = df.sort_values(by=0)
df_s
```

```python
df_sd = df_s.drop(3, axis=1)
```

```python
dmat = np.zeros((943, 1682))
for i in range(df_sd.shape[0]):
    dmat[df_sd.iloc[i, 0] - 1, df_sd.iloc[i, 1] - 1] = df_sd.iloc[i, 2]
```

```python
df_mat = pd.DataFrame(dmat)
df_mat
```

```python
df_mat.to_csv("data_mat.csv", header=None)
```
